```python
#importing necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#Read the csv file which I've taken from kaggle
df = pd.read_csv("C:/Users/BALA/OneDrive - University of Hertfordshire (1)/output.csv")
df
```

```python
# Describing the data for better understanding
desc_stats = df.describe()
print("Summary Statistics:\n", desc_stats)
```

```python
#Summary of the we've data taken
df.info()
```

```python
# Checking the number of null values in each column
null_values = df.isnull().sum()
print(null_values)
```

```python
# Removing columns which have mostly empty values, which doesn't make any difference to the data
df = df.drop(['engine', 'owner', 'location', 'mileage', 'power', 'seats', 'type'], axis=1)
```

```python
# Dropping null values
df = df.dropna()
```

```python
# Checking df after cleaning
df
```

```python
# Calculate correlation matrix with only numeric columns
correlation_matrix = df.select_dtypes(include=['float64', 'int64']).corr()
print("\nCorrelation Matrix:\n", correlation_matrix)
```

```python
#Selecting top 5 brands as per sales
top_brands = df['brand'].value_counts().nlargest(5)
print(top_brands)
```

```python
# Function to plot a bar chart for top 5 car brands
def plot_top_brands(df):
    plt.figure(figsize=(10, 6))
    sns.barplot(x=top_brands.index, y=top_brands.values, hue=top_brands.index, palette="viridis", dodge=False, legend=False)
    plt.title("Top 5 Most Common Car Brands")
    plt.xlabel("Brands")
    plt.ylabel("Count")
    plt.savefig("barplot.png")
    plt.show()

# Call the function to plot
plot_top_brands(df)
```

```python
# Function to plot the line graph
def plot_age_vs_top_brands(df):
    # Calculating the top 10 most common brands
    top_brands = df['brand'].value_counts().nlargest(10).index.tolist()
    
    # Filter the DataFrame for only top brands
    filtered_df = df[df['brand'].isin(top_brands)]
    
    # Calculate the average age for each brand
    average_age = filtered_df.groupby('brand')['age'].mean().reset_index()
    
    # Sort by age for better line plotting
    average_age = average_age.sort_values(by='brand')
    
    # Plotting the line graph
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=average_age, x='brand', y='age', marker='o') 
    plt.title("Average Age of Top 10 Car Brands")
    plt.xlabel("Car Brand")
    plt.ylabel("Average Age (Years)")
    plt.xticks(rotation=45)
    plt.grid()
    plt.savefig("linegraph.png")
    plt.show()

# Call the function to plot
plot_age_vs_top_brands(df)
```

```python
# Function to plot correlation heatmap
def plot_correlation_heatmap(correlation_matrix):
    plt.figure(figsize=(10, 6))
    sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", square=True)
    plt.title("Correlation Heatmap of Numeric Features")
    plt.savefig("Heatmap.png")
    plt.show()

# Call the function to plot the heatmap
plot_correlation_heatmap(correlation_matrix)
```
